```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")

# Display first 5 rows
df.head()
```

```python
# Check all values
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

```python
# finding number of rows and columns
print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
# viewing the data types of columns
df.dtypes
```

```python
# Converting relevant columns to numeric just in case there are strings
df['Exam_Score'] = pd.to_numeric(df['Exam_Score'], errors='coerce')
df['Hours_Studied'] = pd.to_numeric(df['Hours_Studied'], errors='coerce')

df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Hours_Studied'], bins=20, kde=True)
plt.title("Distribution of Hours Studied")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Exam_Score'], bins=20, color='green', kde=True)
plt.title("Distribution of Exam Scores")
plt.show()
```

```python
parental_scores = df.groupby('Parental_Involvement')['Exam_Score'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
parental_scores.plot(kind='bar', color='skyblue')
plt.title("Average Exam Score by Parental Involvement")
plt.ylabel("Average Score")
plt.show()

parental_scores
```

```python
motivation_scores = df.groupby('Motivation_Level')['Exam_Score'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
motivation_scores.plot(kind='bar', color='salmon')
plt.title("Average Exam Score by Motivation Level")
plt.ylabel("Average Score")
plt.show()

motivation_scores
```

```python
# Viewing the highest previous scores recorded
top_previous = df.sort_values(by='Previous_Scores', ascending=False).head(10)
top_previous[['Previous_Scores', 'Exam_Score']]
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df, alpha=0.5)
plt.title("Hours Studied vs. Exam Score")
plt.show()
```

```python
plt.figure(figsize=(10,8))
# Selecting only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
# plt.title("Correlation Matrix of Student Factors")
plt.show()
```
